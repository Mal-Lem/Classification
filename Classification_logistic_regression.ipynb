```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.linear_model import SGDClassifier
```

```python
np.random.seed(1)
X , y = make_classification(n_samples=400 , n_features=2 , n_redundant=0 , n_clusters_per_class=2)
```

```python
#init du model
plt.figure(num = None , figsize=(8,6))
plt.scatter(X[:,0],X[:,1] , marker='o' , c=y , edgecolors='k' )
plt.xlabel('X0')
plt.xlabel('X1')
X.shape
```

```python
#definition et entrainement du model  
model = SGDClassifier(max_iter= 1000 , eta0=0.001 , loss='log')
model.fit(X , y)
print('score  : ', model.score(X,y))
```

```python
h = .02
colors = "bry"
x_min , x_max = X[:,0].min() - 1 , X[:,0].max() + 1
y_min , y_max = X[:,1].min() - 1 , X[:,1].max() + 1
xx , yy = np.meshgrid(np.arange(x_min,x_max,h) , np.arange(y_min , y_max , h))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
cs = plt.contourf(xx , yy , Z , cmap = plt.cm.Paired)
plt.axis('tight')
for i, color in zip (model.classes_ , colors):
    idx = np.where(y == i)
    plt.scatter(X[idx,0] , X[idx,1] , c = color , cmap=plt.cm.Paired , edgecolors='black', s=20)
```
